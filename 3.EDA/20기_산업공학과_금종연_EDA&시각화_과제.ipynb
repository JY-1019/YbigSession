```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
# notebook을 실행한 브라우저에서 바로 그림을 볼 수 있게 해주는 코드. 버전에 따라 필요 없을 수도 있습니다.
import seaborn as sns
```

```python
df = pd.read_csv('airbnb_listings.csv', index_col=0)
```

### 2) 결측치 확인

- #### <font color='orange'> Q2. 각 컬럼의 결측치를 확인하고, 결측치가 많은 순서대로 결측치 비율과 함께 표로 나타내보세요.
<font color='orange'> (결측치 0개인 컬럼은 나타내지 마세요)

```python
df.isnull().sum().to_frame()
```

```python
df_num = df.isnull().sum()
df_num = df_num[df_num>0].sort_values(ascending = False)
df_num = pd.DataFrame(df_num, columns = ['Null'])
df_num['Percentage'] = round(df_num['Null']/len(df),2)
df_num
```

- #### <font color='orange'> Q3. 결측치가 있는 컬럼과 결측치 비율을 보기쉽게 그래프로 나타내 보세요.

```python
# Q3
plt.figure(figsize = (12,6))
sns.barplot(x = df_num.index, y = df_num['Percentage'], data = df_num)
plt.xticks(rotation=90)
plt.show()
```

- #### <font color='orange'> Q4. 결측치가 적은 beds 컬럼을 좀 더 자세히 살펴보겠습니다. 
  #### <font color='orange'> 'beds' 컬럼이 결측치인 데이터의 상위 5개를 출력해주세요.

```python
# Q4
df[df['beds'].isnull()].head()
```

---
### 3) 범주형 변수 확인
<font color='gray'> - 편의를 위해 결측치가 있는 컬럼을 모두 제거하고 진행하겠습니다.

```python
drop_col = ['jurisdiction_names', 'license', 'square_feet', 'monthly_price',
       'weekly_price', 'host_response_rate', 'host_response_time',
       'security_deposit', 'host_acceptance_rate', 'review_scores_location',
       'review_scores_value', 'review_scores_checkin',
       'review_scores_accuracy', 'review_scores_communication',
       'review_scores_cleanliness', 'review_scores_rating',
       'reviews_per_month', 'beds']
```

```python
df = df.drop(columns=drop_col)
```

- #### 각 컬럼별 변수 타입 확인

```python
df.info()
```

- #### <font color='orange'> Q5. Room Type 에는 어떤 것들이 있는지 살펴보겠습니다. 
#### <font color='orange'> room_type의 종류별 갯수에 대한 그래프를 그려주세요.

```python
# Q5
sns.countplot(x = 'room_type', data = df)
```

- #### <font color='orange'> Q6. 위 그래프의 세로축이 갯수가 아닌 비율을 나타내도록 재구성해보세요.

```python
# Q6
plt.figure(figsize=(10,5))
sns.barplot(x = df['room_type'].value_counts().index, y = df['room_type'].value_counts()/len(df))
plt.show()
```

- #### <font color='orange'> Q7. 각 지역별로 room_type의 비율이 어땠는 지 알고 싶습니다.
#### <font color='orange'> 아래 그래프와 같이 각 지역별로 room_type의 갯수가 어땠는지 보여주세요.    

```python
# Q7
plt.figure(figsize=(12,6))
sns.countplot(x = 'neighborhood', data = df, hue = 'room_type')
plt.show()
```

- #### <font color='orange'> Q8. 각 지역별로 가격의 분포가 어떤지 알고 싶습니다.
#### <font color='orange'> 아래 그래프와 같이 각 지역별로 가격의 분포를 보여주세요.

```python
# Q8
# plt.figure() 여기부턴 직접 사이즈 조절해보세요
fig = plt.figure(figsize=(15,15))
ax1 = fig.add_subplot(2,1,1)
ax2 = fig.add_subplot(2,1,2)
sns.stripplot(x='neighborhood', y='price', data=df, jitter=0.2, ax=ax1)
sns.boxplot(x='neighborhood', y='price', data=df, ax=ax2)
plt.ylim(0,1000)
plt.show()
```

- #### <font color='orange'> Q9. 위 그래프들로부터 알 수 있는 사실은 무엇인가요? 그 이유는 무엇이라고 생각하시나요?

- Manhatten이나 Brooklyn에 다른 방들에 비해 유독 비싼 방들이 많다. 또한 boxplot과 stripplot을 비교하면 알 수 있듯이 역시 빈부격차의 땅 미국답게 IQR을 적용했을 경우 이상치로 분류될 만큼의 비싼 방들이 굉장히 많이 있다. 따라서 price 데이터를 통해 인사이트를 얻으려면 이상치에 대해 정교한 분석이 필요해 보인다.

- Manhatten의 경우, 이상하게 Private Room 보다 Entire home이 더 비싼데 이는 땅값과 관련있을 수도 있고 아파트 형태로 된 방이 많아서 그런 것일 수도 있다. 



```python
df_hotel = df[df['room_type']=='Hotel room']
df_hotel
plt.figure(figsize=(5,5))
sns.stripplot(x='neighborhood', y='price', data=df_hotel)
plt.ylim(0,2000)
plt.show()
```
